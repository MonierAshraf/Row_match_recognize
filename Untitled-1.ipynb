```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from matplotlib.patches import Rectangle

# Set style
plt.style.use('default')
sns.set_palette("husl")

# Create figure with subplots
fig = plt.figure(figsize=(16, 12))


# Data
cache_modes = ['fifo', 'lru', 'none']
avg_times = [4.009, 3.433, 3.778]
memory_usage = [0.00, 0.21, 1.90]
hit_rates = [90.9, 90.9, 0]
improvements = [-6.1, 9.2, 0]

# Scenario data
scenarios = ['Basic patterns,\nsmall dataset', 'Complex\npatterns', 'Complex patterns,\nlarge dataset']
fifo_times = [1.468, 3.019, 7.540]
lru_times = [1.444, 3.205, 5.649]
none_times = [1.416, 3.113, 6.806]

# Scalability data
dataset_sizes = [1000, 2000, 4000]
none_scale = [1.5, 3.0, 6.5]
fifo_scale = [1.6, 3.2, 7.2]
lru_scale = [1.4, 2.8, 5.5]

# 1. Average Execution Time
ax1 = plt.subplot(3, 4, 1)
colors = ['#ff7f7f', '#7fbfff', '#ff9999']
bars = ax1.bar(cache_modes, avg_times, color=colors, alpha=0.8)
ax1.set_title('Average Execution Time', fontweight='bold', pad=10)
ax1.set_ylabel('Time (seconds)')
ax1.set_ylim(0, 4.5)

# Add value labels on bars
for bar, time in zip(bars, avg_times):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 0.05, f'{time:.3f}s',
             ha='center', va='bottom', fontweight='bold')

# 2. Performance vs Baseline
ax2 = plt.subplot(3, 4, 2)
colors_perf = ['#ff7f7f', '#7fbfff']
improvements_data = [-6.1, 9.2]
bars2 = ax2.bar(['FIFO', 'LRU'], improvements_data, color=colors_perf, alpha=0.8)
ax2.set_title('Performance vs Baseline', fontweight='bold', pad=10)
ax2.set_ylabel('Improvement (%)')
ax2.axhline(y=0, color='red', linestyle='--', alpha=0.7)
ax2.set_ylim(-8, 10)

# Add value labels
for bar, imp in zip(bars2, improvements_data):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + (0.3 if height > 0 else -0.5), 
             f'{imp:.1f}%', ha='center', va='bottom' if height > 0 else 'top', fontweight='bold')

# 3. Cache Hit Rates (Pie Chart)
ax3 = plt.subplot(3, 4, 3)
sizes = [50, 50]
colors_pie = ['#7fdfbf', '#7fbfff']
labels = ['FIFO', 'LRU']
wedges, texts, autotexts = ax3.pie(sizes, labels=labels, colors=colors_pie, autopct='%1.1f%%',
                                   startangle=90, textprops={'fontweight': 'bold'})
ax3.set_title('Cache Hit Rates', fontweight='bold', pad=10)

# Add 90.9% annotation
ax3.text(0, -1.3, '90.9%', ha='center', va='center', fontsize=12, fontweight='bold')

# 4. Memory Usage
ax4 = plt.subplot(3, 4, 4)
colors_mem = ['#7fdfbf', '#7fbfff', '#ff9999']
bars4 = ax4.bar(cache_modes, memory_usage, color=colors_mem, alpha=0.8)
ax4.set_title('Memory Usage', fontweight='bold', pad=10)
ax4.set_ylabel('Memory Increase (MB)')
ax4.set_ylim(0, 2.0)

# Add value labels
for bar, mem in zip(bars4, memory_usage):
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height + 0.05, f'{mem:.2f}MB',
             ha='center', va='bottom', fontweight='bold')

# Add baseline annotation
ax4.text(2.3, 1.5, '1.90MB', ha='left', va='center', fontsize=10, fontweight='bold')

# 5. Performance by Scenario
ax5 = plt.subplot(3, 4, (5, 6))
x = np.arange(len(scenarios))
width = 0.25

bars5_1 = ax5.bar(x - width, none_times, width, label='NONE', color='#ff9999', alpha=0.8)
bars5_2 = ax5.bar(x, fifo_times, width, label='FIFO', color='#7fdfbf', alpha=0.8)
bars5_3 = ax5.bar(x + width, lru_times, width, label='LRU', color='#7fbfff', alpha=0.8)

ax5.set_title('Performance by Scenario', fontweight='bold', pad=10)
ax5.set_ylabel('Execution Time (seconds)')
ax5.set_xlabel('Test Scenario')
ax5.set_xticks(x)
ax5.set_xticklabels(scenarios, rotation=45, ha='right')
ax5.legend()
ax5.set_ylim(0, 8)

# 6. Scalability Analysis
ax6 = plt.subplot(3, 4, (7, 8))
ax6.plot(dataset_sizes, none_scale, 'o-', color='#ff9999', label='NONE', linewidth=2, markersize=8)
ax6.plot(dataset_sizes, fifo_scale, 's-', color='#7fdfbf', label='FIFO', linewidth=2, markersize=8)
ax6.plot(dataset_sizes, lru_scale, '^-', color='#7fbfff', label='LRU', linewidth=2, markersize=8)

ax6.set_title('Scalability Analysis', fontweight='bold', pad=10)
ax6.set_ylabel('Execution Time (seconds)')
ax6.set_xlabel('Dataset Size (records)')
ax6.legend()
ax6.grid(True, alpha=0.3)
ax6.set_ylim(0, 8)

# 7. Summary Table
ax7 = plt.subplot(3, 4, (9, 12))
ax7.axis('off')

# Create table data
table_data = [
    ['Cache Mode', 'Avg Time', 'Memory', 'Hit Rate'],
    ['NONE', '3.778s', '1.90MB', 'N/A'],
    ['FIFO', '4.009s', '0.00MB', '90.9%'],
    ['LRU', '3.433s', '0.21MB', '90.9%']
]

# Create table
table = ax7.table(cellText=table_data[1:], colLabels=table_data[0],
                  cellLoc='center', loc='center', bbox=[0, 0.6, 1, 0.3])
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1, 2)

# Style table
for i in range(len(table_data[0])):
    table[(0, i)].set_facecolor('#4a4a4a')
    table[(0, i)].set_text_props(weight='bold', color='white')

# Color code rows
table[(1, 0)].set_facecolor('#ffe6e6')  # NONE
table[(2, 0)].set_facecolor('#e6f3ff')  # FIFO  
table[(3, 0)].set_facecolor('#e6f3ff')  # LRU

# Add performance summary
summary_text = "Performance Summary: LRU delivers 9.2% improvement with 90.9% cache efficiency"
ax7.text(0.5, 0.3, summary_text, ha='center', va='center', fontsize=12, 
         fontweight='bold', bbox=dict(boxstyle="round,pad=0.3", facecolor='lightblue', alpha=0.7))

# Adjust layout
plt.tight_layout()
plt.subplots_adjust(top=0.92, hspace=0.3, wspace=0.3)

# Show plot
plt.show()

# Optional: Save the figure
# plt.savefig('performance_dashboard.png', dpi=300, bbox_inches='tight')
```
